# Anlisis


Se calcula la Brecha entre el dlar oficial y el dlar blue.
Agregamos la columna Brecha en Porcentaje, segun la frmula:

((dolar_blue - dolar_of) / dolar_of) * 100

se redondea a dos decimales para facilitar la visualizacin.

```python
dolar_blue_of['brecha'] = ((dolar_blue_of['valor_blue'] - dolar_blue_of['valor_oficial']) / dolar_blue_of['valor_oficial']) * 100
#redondear brecha a dos decimales
dolar_blue_of['brecha'] = dolar_blue_of['brecha'].round(2)
```

```python
# se muestra los ltimos registros del nuevo dataframe.
dolar_blue_of.tail()
```

```python
#resertear el indice del dataframe
dolar_blue_of.reset_index(inplace=True)
```

```python
# se muestra la informacin del nuevo dataframe.
dolar_blue_of.info()
```

```python
#se muestra los primeros registros del nuevo dataframe.
dolar_blue_of.head()
```

```python
dolar_blue_of['fecha'] = pd.to_datetime(dolar_blue_of['fecha'])
dolar_blue_of['mes'] = dolar_blue_of['fecha'].dt.month 
dolar_blue_of['ao'] = dolar_blue_of['fecha'].dt.year
dolar_blue_of['semana'] = dolar_blue_of['fecha'].dt.week 
dolar_blue_of['dia'] = dolar_blue_of["fecha"].dt.day_name()
```

```python
#cambiar indice por fecha
dolar_blue_of.set_index('fecha', inplace=True, drop= True)
```

```python
#reordenar las columnas del dataframe
dolar_blue_of = dolar_blue_of[['valor_blue', 'valor_oficial', 'brecha','dia', 'semana','mes',  'ao']]
```

Vemos como queda ordenada la informacin

```python
dolar_blue_of.tail()
```

```python
#seleccionamos los datos de los ultimos 365 registros 
analisis_anual = dolar_blue_of.tail(365)
```

```python
analisis_anual.tail()
```

```python
#graficar anlisis anual de la brecha
analisis_anual.plot(y='brecha', figsize=(10, 6), kind = 'line', title = 'Anlisis Anual de la Brecha Dolar Blue/Dolar Oficial')
plt.xlabel("fecha")
plt.ylabel("Brecha %")
plt.show()
```

```python
print(analisis_anual.columns)
```

```python
# Seleccionar los datos correspondientes a los ltimos dos aos
datos_ultimos_dos_anos = analisis_anual.loc[analisis_anual['ao'] >= 2021]

# Agrupar los datos por mes y calcular la media del valor_blue y valor_oficial
datos_mensuales = datos_ultimos_dos_anos.groupby(['ao', 'mes']).mean()[['valor_blue', 'valor_oficial']]

# Crear un nuevo dataframe con los datos obtenidos
evolucion_mensual = pd.DataFrame({
    'valor_blue': datos_mensuales['valor_blue'].values,
    'valor_oficial': datos_mensuales['valor_oficial'].values
}, index=datos_mensuales.index)

# Graficar la evolucin mensual del valor_blue y valor_oficial
evolucion_mensual.plot(figsize=(10,6), title='Evolucin Mensual de la Cotizacin del Dlar Blue y Oficial en los ltimos 2 Aos')
plt.xlabel('Mes y Ao')
plt.ylabel('Cotizacin')
plt.show()
```
